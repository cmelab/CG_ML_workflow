```python
import h5py
import numpy as np
import pandas as pd
```

```python
rigid_traj_path = '/home/marjanalbooyeh/projects/ellipsoid_sim/PPS_rigid/logs/pps_20_5_aug18/rigid_traj.h5'
```

```python
f = h5py.File(rigid_traj_path, 'r')
```

### 2. Get data from frames

```python
columns = ['net_force', 'net_torque', 'orientation', 'position']
```

```python
traj_df = pd.DataFrame()
for frame in f.keys():
    dset = dict(f[frame])
    dset = {key: np.array(value) for key, value in dset.items()}
    dset['timestep'] = f[frame].attrs['timestep']
    traj_df = traj_df.append(dset, ignore_index=True)
```

```python
f.close()
```

```python
traj_df['position'].shape
```

```python
traj_df.head()
```

```python
#shuffle order of frames
traj_df = traj_df.sample(frac=1).reset_index(drop=True)
```

```python
traj_df.head()
```

### 3. Split dataset & Normalize 

```python
test_frac = 0.1
val_frac = 0.1
```

```python
dataset_len = traj_df.shape[0]
test_len = int(dataset_len * test_frac)
val_len = int(dataset_len * val_frac)

test_df = traj_df.loc[:test_len]
train_val_df = traj_df.loc[test_len:]
val_df = pd.DataFrame(columns=columns)
train_df = pd.DataFrame(columns=columns)
```

```python
stats = {
    'force_mean': np.asarray(list(train_val_df['net_force'])).mean(axis=0),
    'force_std' : np.asarray(list(train_val_df['net_force'])).std(axis=0),
    'torque_mean' : np.asarray(list(train_val_df['net_torque'])).mean(axis=0),
    'torque_std' : np.asarray(list(train_val_df['net_torque'])).std(axis=0)
}
```

```python
columns = list(set(traj_df.columns) - set(['timestep']))
```

```python
for col in columns:
    train_col = np.asarray(list(train_val_df[col]))
    test_col = np.asarray(list(test_df[col]))
    mean = train_col.mean(axis=0)
    std = train_col.std(axis=0)
    train_col = ((train_col - mean) / std)
    val_df[col] = train_col[:val_len].tolist()
    train_df[col] = np.array(((train_col[val_len:] - mean) / std).tolist())
    test_df.loc[:, col] = ((test_col - mean) / std).tolist()
```
